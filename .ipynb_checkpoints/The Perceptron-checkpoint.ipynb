# The Perceptron
* artificial neuron - receives input => does computation on it => produces an output

```python
import numpy as np
import matplotlib.pyplot as plt
import neurolab as nl
plt.style.use('seaborn')
plt.rcParams['axes.facecolor'] = 'darkgray'
```

```python
num_dataset = np.array([[0.22,0.34,0],[0.21,0.37,0],[0.25,0.31,0],[0.76,0.19,1],[0.84,0.14,1]])
```

```python
X = num_dataset[:,:2]
y = num_dataset[:,2].reshape((num_dataset.shape[0],1))
```

```python
print(X.shape)
print(y.shape)
```

```python
plt.scatter(X[:,0],X[:,1],c=y)
plt.xlabel("Dimension 1")
plt.ylabel("Dimenision 2")
plt.title("Input Data")
plt.show()
```

```python
dim1_min,dim1_max,dim2_min,dim2_max = 0,1,0,1
```

```python
num_output = y.shape[1]
print(num_output)
```

```python
dim1 = [dim1_min,dim1_max]
dim2 = [dim2_min,dim2_max]
```

```python
perceptron = nl.net.newp([dim1,dim2],num_output)
```

```python
error_progress = perceptron.train(X,y,epochs=100,show=20,lr=0.03)

plt.plot(error_progress)
plt.xlabel("Number of Epochs")
plt.ylabel("Training Error")
plt.title("Training Error Progress")
plt.grid('on')
```

```python
perceptron.sim([[0.81,0.23]])
```
